```python
Q1 = df['Sale Price'].quantile(0.25)
Q2 = df['Sale Price'].quantile(0.50)
Q3 = df['Sale Price'].quantile(0.75)
IQR = Q3 - Q1
print(Q1)
print(Q2)
print(Q3)
print(IQR)
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(lower_bound)
print(upper_bound)
df['Sale Price']=df['Sale Price'].clip(lower_bound,upper_bound)
df.boxplot('Sale Price')
```

```python
outliers=[]
for row in df['Sale Price']:
  if (row<lower_bound) or  (row>upper_bound):
    outliers.append(row)
print(outliers)

```
