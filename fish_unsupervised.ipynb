```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

df.head()
```

```python
for col in df.columns:
  sns.boxplot(df[col])
  plt.show()
```

```python
df.head()
```

```python
feature_cols =['242.0','23.2','25.4','30.0']
for col in feature_cols:
  Q1 = df[col].quantile(0.25)
  Q2 = df[col].quantile(0.50)
  Q3 = df[col].quantile(0.75)
  IQR = Q3 - Q1
  lb = Q1 - 1.5*IQR
  ub = Q3 + 1.5*IQR
  df[col] = df[col].clip(lb,ub)
for col in feature_cols:
  sns.boxplot(df[col])
  plt.show()
```

```python
df['Bream'].unique()
```

```python
from sklearn.cluster import KMeans
import pandas as pd

numerical_df = df.select_dtypes(include=['number'])

wcss = []
cluster_range = range(1, 11)

for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(numerical_df)
    wcss.append(kmeans.inertia_)
```

```python
pred_clusters = kmeans.predict(numerical_df)
pred_clusters
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['13.4'],df['23.2'],c=pred_clusters,cmap='rainbow',s=50)
centroids = kmeans.cluster_centers_
plt.title('K Means Clustering')
plt.xlabel('13.4')
plt.ylabel('23.2')
```

```python
from sklearn.cluster import AgglomerativeClustering
agglo_clustering = AgglomerativeClustering(metric='euclidean', linkage='ward')
y_agglo = agglo_clustering.fit_predict(numerical_df )
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['13.4'],df['23.2'],c=y_agglo,cmap='rainbow',s=50)
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:,0],centroids[:,1],c='black',s=200,alpha=0.5)
plt.title('Agglomerative Clusturing')
plt.xlabel('13.4')
plt.ylabel('23.2')
```

```python
from sklearn.metrics import silhouette_score
sillhouette_score = silhouette_score(numerical_df,y_agglo)
sillhouette_score
```

```python
silh_score = []
for i in range(2,11):
  agglo_clustering = AgglomerativeClustering(n_clusters=i,metric='euclidean', linkage='ward')
  agglo_labels = agglo_clustering.fit_predict(numerical_df)
  score = silhouette_score(numerical_df,agglo_labels)
  silh_score.append(score)
  print(f'Silhouette score for {i} clusters is {score}')
```

```python
plt.plot(range(2,11),silh_score,marker = 'o')
plt.title('Silhouette Score')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.show()
```
