```python
from sklearn.impute import SimpleImputer
import numpy as np
```

```python
null_colums = df[['satisfaction', 'evaluation', 'number_of_projects',
       'average_montly_hours','department']]
```

```python
imputer = SimpleImputer(strategy='most_frequent')
null_colums = imputer.fit_transform(null_colums)
```

```python
null_colums = pd.DataFrame(null_colums, columns=['satisfaction', 'evaluation', 'number_of_projects',
       'average_montly_hours','department'])

null_colums.isna().sum()
```
